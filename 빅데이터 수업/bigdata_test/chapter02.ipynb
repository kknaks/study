```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
print(type(iris))

keys = iris.keys()
print(keys)
```

```python
print(iris.feature_names)
print(iris.target_names)
print(iris.data)
print(iris.target)
```

```python
import pandas as pd

iris_df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
iris_df['label']  = iris.target
iris_df
```

```python
from sklearn.model_selection import train_test_split

iris_data = iris.data
iris_label= iris.target

x_train, x_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size= 0.2, random_state=11)

print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier

df_clf = DecisionTreeClassifier(random_state=11)
df_clf.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

pred = df_clf.predict(x_test)
ac_score = accuracy_score(y_test, pred)

ac_score
```

# 스케일링


```python
from sklearn.datasets import load_iris
import pandas as pd 

iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data = iris_data, columns= iris.feature_names)

print('평균\n',iris_df.mean())
print('분산\n',iris_df.var())
```

```python
#스케일링
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)
iris_df_scaled = pd.DataFrame(data = iris_scaled, columns= iris.feature_names)

print('평균\n',iris_df_scaled.mean())
print('분산\n',iris_df_scaled.var())
iris_df_scaled
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler2 = MinMaxScaler()
scaler2.fit(iris_df)
iris_scaled2 = scaler2.transform(iris_df)
iris_df_scaled2 = pd.DataFrame(data  = iris_scaled2, columns= iris.feature_names)

print('평균\n',iris_df_scaled2.min())
print('분산\n',iris_df_scaled2.max())
iris_df_scaled2
```

```python
#스케일링 주의 사항
from sklearn.preprocessing import MinMaxScaler
import numpy as np 

train_array = np.arange(0,11).reshape(-1,1)
test_array = np.arange(0,6).reshape(-1,1)

scaler = MinMaxScaler()
scaler.fit(train_array)
train_scaled =scaler.transform(train_array)

print('원본 트레인 데이터 \n', np.round(train_array.reshape(-1),2))
print('원본 트레인 데이터 \n', np.round(train_scaled.reshape(-1),2))

test_scaled = scaler.transform(test_array)

print('test 트레인 데이터 \n', np.round(test_array.reshape(-1),2))
print('test 트레인 데이터 \n', np.round(test_scaled.reshape(-1),2))


scaler.fit(test_array)
test_scaled = scaler.transform(test_array)

print('test2 트레인 데이터 \n', np.round(test_array.reshape(-1),2))
print('test2 트레인 데이터 \n', np.round(test_scaled.reshape(-1),2))
```
