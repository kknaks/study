```python
import pandas as pd
import requests
from tqdm import tqdm
import geopandas
```

```python
#인풋데이터 불러오기
input_data = pd.read_csv('./df2.csv')
input_data = input_data.reset_index(drop=True)
```

```python
#위도(lat),경도(lng)불러오는 함수
def get_lat_lng(address):
    #카카오톡 api 불러오기
    url = f'https://dapi.kakao.com/v2/local/search/address.json?query={address}'
    headers = {
    ## 여러분의 카카오 API의 REST API키를 아래 예시와 같이 입력해주세요
    ## "Authorization": "KakaoAK REST API키 입력 gogo"}
    "Authorization": "KakaoAK 1358366e6234ef1b494f623fa9b7fd03"}
    response = requests.get(url, headers = headers).json()
    try : 
        #위경도
        Lat = response.get('documents')[0].get('y')
        Lng = response.get('documents')[0].get('x')
    
    except : 
        Lat = 'error'
        Lng = 'error'

    return Lat, Lng

```

```python
df = input_data
```

```python
df_test = df.iloc[:100,:]
```

```python
df_test['Lat_Lng'] = df_test.지번주소.apply(lambda x : get_lat_lng(x)) 
```

```python
df['Lat_Lng'] = df.지번주소.apply(lambda x : get_lat_lng(x)) 
```

```python
df.to_csv('./output.csv', index=False, encoding='utf-8-sig')
```
