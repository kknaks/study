```python
#변수분리
COL_DEL=['ID']
COL_NUM=['Customer_care_calls','Cost_of_the_Product','Prior_purchases','Discount_offered','Weight_in_gms']
COL_CAT=['Warehouse_block','Mode_of_Shipment','Customer_rating','Product_importance','Gender']
COL_Y=['Reached.on.Time_Y.N']
```

```python
#변수분리
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(X_train[COL_NUM+COL_CAT],
                                            y_train[COL_Y].values.ravel(),
                                            test_size =0.3)
```

```python
#정규화
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
std.fit(X_train[COL_NUM])

X_tr[COL_NUM] = std.transform(X_tr[COL_NUM])
X_val[COL_NUM] = std.transform(X_val[COL_NUM])
X_test[COL_NUM] = std.transform(X_test[COL_NUM])
```

```python
#범주화
from sklearn.preprocessing import LabelEncoder

X = pd.concat([X_train, X_test])
for col in COL_CAT:
    le = LabelEncoder()
    le.fit(X[col])
    X_tr[col] = le.transform(X_tr[col])
    X_val[col] = le.transform(X_val[col])
    X_test[col] = le.transform(X_test[col])        
    print(col)
    print(le.classes_)
    print('\n')
```

## 2.모형생성

```python
#랜덤포레스트
from sklearn.ensemble import RandomForestClassifier
modelRF = RandomForestClassifier(random_state = 123)
modelRF.fit(X_tr,y_tr)


#부스팅
from xgboost import XGBClassifier
modelXG = XGBClassifier(random_state = 123)
modelXG.fit(X_tr,y_tr)
```

```python
y_pred_RF = modelRF.predict_proba(X_val)[:,1]
y_pred_XG = modelXG.predict_proba(X_val)[:,1]

from sklearn.metrics import roc_auc_score
score_rf = roc_auc_score(y_val, y_pred_rf)
score_xg = roc_auc_score(y_val, y_pred_xg)

print(score_rf,score_xg)
```

```python
#
def get_score(model, X_tr, X_val, y_tr, y_val):
    y_pred_tr = model.predict_proba(X_tr)[:,1]
    y_pred_val = model.predict_proba(X_val)[:,1]

    score_tr = roc_auc_score(y_tr, y_pred_tr)
    score_val = roc_auc_score(y_val, y_pred_val)

    return f'train : {score_tr}, valid : {score_val }'
```

```python
print(f'RandomForest --', get_score(modelRF, X_tr, X_val, y_tr, y_val))
print(f'XGBoosting --', get_score(modelXG, X_tr, X_val, y_tr, y_val))
```

```python
#하이퍼파라미터 튜닝
modelXG2 = XGBClassifier(n_estimators = 100, min_child_weight=6,max_depth =3, min_split_loss=10,subsample=1)
modelXG2.fit(X_tr,y_tr)
print(f'XGBoosting --', get_score(modelXG2, X_tr, X_val, y_tr, y_val))

modelXG3 = XGBClassifier(n_estimators = 50,min_child_weight=7,max_depth =5, min_split_loss=10,subsample=1)
modelXG3.fit(X_tr,y_tr)
print(f'XGBoosting --', get_score(modelXG3, X_tr, X_val, y_tr, y_val))

modelXG4 = XGBClassifier(n_estimators = 50,min_child_weight=8,max_depth =3, min_split_loss=10,subsample=1)
modelXG4.fit(X_tr,y_tr)
print(f'XGBoosting --', get_score(modelXG4, X_tr, X_val, y_tr, y_val))

modelXG5 = XGBClassifier(n_estimators = 100,min_child_weight=9,max_depth =3, min_split_loss=10,subsample=1)
modelXG5.fit(X_tr,y_tr)
print(f'XGBoosting --', get_score(modelXG4, X_tr, X_val, y_tr, y_val))
```
